```python
from pyspark.streaming import StreamingContext
from pyspark.sql import Row
import time
```

```python
def func_1(time, rdd):
    try:
        df = spark.createDataFrame(rdd.map(\
        lambda row: Row(time=time, word=row[0], count=row[1])))
        df.write.format("org.apache.spark.sql.cassandra")\
        .options(table="assignment_count1", keyspace="streaming")\
        .save(mode="append")
    except:
        pass
```

```python
#1. Show total number of counts (on each site) by vehicle class. (10 marks) 

lines = ssc.textFileStream("file:///home/madhavi/Assignment3/Streaming/")
vehicle_counter_DF = lines.map(lambda line:line.split(","))
vehicles_count = vehicle_counter_DF.map(lambda x: ((x[0],x[14]),1))
vehicles_count = vehicles_count.reduceByKey(lambda a,b : a+b)

vehicles_count.foreachRDD(func_1)
vehicles_count.pprint()

ssc.start()
time.sleep(300)
ssc.stop(stopSparkContext=False)
```

```python
# 2. Computetheaveragespeed(oneachsite)byvehicleclass. (10 marks) 

lines = ssc.textFileStream("file:///home/madhavi/Assignment3/streaming/")
vehicle_counter_DF = lines.map(lambda line:line.split(","))

vehicle_avg_speed = vehicle_counter_DF.map(lambda x: ((x[18]),1))
vehicle_avg_speed = vehicle_avg_speed.reduceByKey(lambda a,b : (a[0]+b[0], a[1]+b[1]))
vehicle_avg_speed = vehicle_avg_speed.mapValues(lambda x: x[0])
vehicle_avg_speed.foreachRDD(func_2)
vehicle_avg_speed.pprint()

ssc.start()
time.sleep(300)
ssc.stop(stopSparkContext=False)
```

```python
# 3. Find the top3 busiest counter sites in Ireland. (10 marks) 

lines = ssc.textFileStream("file:///home/madhavi/Assignment3/Streaming/")
vehicle_counter_DF = lines.map(lambda line:line.split(","))
busiest_Sites = vehicle_counter_DF.map(lambda x: ((x[0]),1))
busiest_Sites = busiest_Sites.reduceByKey(lambda a,b : a+b)
busiest_Sites=busiest_Sites.transform(
lambda rdd:rdd.sortBy(lambda x:x[1],ascending=False)
)
busiest_Sites.pprint(3)
busiest_Sites.foreachRDD(func_3)

ssc.start()
time.sleep(300)
ssc.stop(stopSparkContext=False)
```
